# Criando pivot tables

```python
import pandas as pd
import numpy as np
```

```python
primary_df = pd.read_csv('datasets\\primary-results.csv')
```

```python
primary_df['candidate'].unique()
```

```python
pd.pivot_table(
    primary_df, 
    index=['state', 'party', 'candidate'], 
    values=['votes'], 
    aggfunc={'votes': np.sum}
)
```

```python
primary_df['rank'] = primary_df.groupby(
    ['county', 'party'])['votes'].rank(ascending=False)
```

```python
primary_df[primary_df['county']=='Los Angeles']
```

```python
primary_df_groupby = primary_df.groupby(['state', 'party', 'candidate']).sum()
del primary_df_groupby['fips']
del primary_df_groupby['fraction_votes']
primary_df_groupby.reset_index(inplace=True)
primary_df_groupby.head(10)
```

```python
primary_df_groupby['rank'] = primary_df_groupby.groupby(['state', 'party'])['votes'].rank(ascending=False)
```

```python
primary_df_groupby.head(7)
```

```python
pd.pivot_table(
    primary_df_groupby, 
    index=['state', 'party', 'candidate'], 
    values=['rank', 'votes']
)
```

```python
primary_df_groupby[primary_df_groupby['rank'] == 1]['candidate'].value_counts()
```

# Visualização 08 - Gráficos de Séries Temporais

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
import matplotlib.dates as dates
import datetime as dt
```

```python
df = pd.read_csv('datasets/ppz-jan-fev-2017.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
def to_date(value):
    return dt.datetime(2017, 1, 1) + dt.timedelta(hours=value)

df['date'] = df['hour'].apply(to_date)
```

```python
df.head()
```

```python
df.tail()
```

```python
del df['hour']
```

```python
df.set_index(['date'], inplace=True)
```

```python
df.head()
```

```python
fig, ax = plt.subplots()
ax.plot_date(df.index.to_pydatetime(), df['views'], 'b-')
ax.xaxis.set_minor_locator(dates.DayLocator(bymonthday=range(5, 32 ,5)))
ax.xaxis.set_minor_formatter(dates.DateFormatter('%d'))
ax.xaxis.grid(True, which="minor")
ax.yaxis.grid()
ax.xaxis.set_major_locator(dates.MonthLocator())
ax.xaxis.set_major_formatter(dates.DateFormatter('%b'))
plt.tight_layout()
plt.show()
```

```python
import locale
locale.setlocale(locale.LC_ALL,'pt-BR')
fig, ax = plt.subplots()
ax.plot_date(df.index.to_pydatetime(), df['views'], 'b-')
ax.xaxis.set_minor_locator(dates.DayLocator(bymonthday=range(5, 32 ,5)))
ax.xaxis.set_minor_formatter(dates.DateFormatter('%d'))
ax.xaxis.grid(True, which="minor")
ax.yaxis.grid()
ax.xaxis.set_major_locator(dates.MonthLocator())
ax.xaxis.set_major_formatter(dates.DateFormatter('%b'))
plt.tight_layout()
plt.show()
```
