# Visualizando os dados com Matplotlib

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
salario = np.array([100, 200, 300, 500, 400, 150])
salario
```

```python
plt.plot(salario)
```

```python
plt.plot(salario, c='Black')
```

```python
plt.plot(salario, c='Black', ls='--')
```

```python
plt.plot(salario, c='Black', ls='--', marker='s')
```

```python
plt.plot(salario, c='Black', ls='--', marker='s', ms=18)
plt.show()
```

## Visualizando dados novamente (aula 54)

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
salarios_marcos = np.array([100, 200, 400, 300])
salarios_gileno = np.array([50, 300, 500, 450])
```

```python
# adicionando label
plt.plot(salarios_marcos, c='Black', ls='--', marker='s', ms=8, label='Salários Marcos')
plt.plot(salarios_gileno, c='Red', ls='--', marker='^', ms=8, label='Salários Gileno')
plt.legend() # colocando legenda
plt.show()
```

```python
# adicionando label
plt.plot(salarios_marcos, c='Black', ls='--', marker='s', ms=8, label='Salários Marcos')
plt.plot(salarios_gileno, c='Red', ls='--', marker='^', ms=8, label='Salários Gileno')
plt.legend(loc='lower right') # posicionando legenda
plt.show()
```

```python
# adicionando label
salarios_pedro = [200, 150, 300, 700]
plt.plot(salarios_marcos, c='Black', ls='--', marker='s', ms=8, label='Salários Marcos')
plt.plot(salarios_gileno, c='Red', ls='--', marker='^', ms=8, label='Salários Gileno')
plt.plot(salarios_pedro, c='Green', ls='--', marker='o', ms=8, label='Salários Pedro')
plt.legend() # posicionando legenda
plt.show()
```
