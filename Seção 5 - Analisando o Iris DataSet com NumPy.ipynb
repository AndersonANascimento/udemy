# Analisando o Iris DataSet com NumPy

```python
import numpy as np
```

```python
data = np.genfromtxt('datasets/iris.data', delimiter=',', usecols=(0,1,2,3))
data
```

```python
# n√∫mero de amostras
len(data)
```

```python
# comprimentos da sepalas das amostras
sepal_lengths = data[:, 0]
sepal_lengths
```

```python
# comprimento das sepalas da Iris Setosa
iris_setosa_sepal_len = data[:50, 0]
iris_setosa_sepal_len
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(iris_setosa_sepal_len, c='Red', ls=':', marker='s', ms=8)
```

```python
# comprimento das sepalas da Iris Versicolor
iris_versi_sepal_len = data[50:100, 0]
iris_versi_sepal_len
```

```python
plt.plot(iris_setosa_sepal_len, c='Red', ls=':', marker='s', ms=8, label='Comp. S√©pala √çris-Setosa')
plt.plot(iris_versi_sepal_len, c='Black', ls=':', marker='o', ms=8, label='Comp. S√©pala √çris-Versicolor')
plt.legend()
plt.show()
```

```python
# comprimento das sepalas da Iris Virginica
iris_virgi_sepal_len = data[100:150, 0]
iris_virgi_sepal_len
```

```python
plt.plot(iris_setosa_sepal_len, c='Red', ls=':', marker='s', ms=8, label='Comp. S√©pala √çris-Setosa')
plt.plot(iris_versi_sepal_len, c='Green', ls=':', marker='o', ms=8, label='Comp. S√©pala √çris-Versicolor')
plt.plot(iris_virgi_sepal_len, c='Blue', ls=':', marker='^', ms=8, label='Comp. S√©pala √çris-Virginica')
plt.legend()
plt.show()
```
